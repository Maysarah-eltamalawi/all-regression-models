## encoding

```python
cat_colum = df.select_dtypes(include = "object").columns
```

# label Encoder


```python
from sklearn.preprocessing import LabelEncoder

# transform data
for col in cat_colum:
  df[col] = LabelEncoder().fit_transform(df[col])
  df[[col,col]] # special syntax to get just these two columns
df.head()
```

# Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np
encoder = OrdinalEncoder()
for col in cat_colum:
  df[col] = encoder.fit_transform((np.array(df[col])).reshape(-1,1))
  df[[col,col]] # special syntax to get just these two columns
  if not df[col].dtypes == "object":
        df[col].fillna(df[col].mean(), inplace = True)
  else:
        df[col].fillna(df[col].mode(), inplace = True)
df.head()
```

# one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

# transform data
for col in cat_colum:
  df[col] = OneHotEncoder().fit_transform(np.array(df[col]).reshape(-1, 1)).toarray()
  df[[col,col]] # special syntax to get just these two columns
df.head()
```

# dummy endoder

```python
# Dummy variable
df_encode = pd.get_dummies(data = df, prefix = 'OHE', prefix_sep='_',
               columns = cat_colum,
               drop_first =True,
              dtype='int8')
df_encode.head()
```

# split the data into train and test 

```python
from sklearn.model_selection import train_test_split

X = df_encode.drop('SalePrice',axis=1) # Independet variable
y = df_encode['SalePrice'] # dependent variable
#split data into train and validation 
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
#split data into train and test 
#train_X, test_X, train_y, test_y = train_test_split(X, y,random_state = 0)
```

# modeling

#1- randomforest 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import math
from math import *

rf = RandomForestRegressor(random_state=42, n_jobs=-1)
rf.fit(train_X,train_y)
rf_pred = rf.predict(val_X)
rf_train_pred = rf.predict(train_X)
print("-"*10,"Test Pred","-"*10)
print("R2: ",r2_score(val_y,rf_pred))
print("MAE: ",mean_absolute_error(val_y,rf_pred))
print("RMSE: ",math.sqrt(mean_squared_error(val_y,rf_pred)))

print("-"*10,"Train Pred","-"*10)
print("R2: ",r2_score(train_y,rf_train_pred))
print("MAE: ",mean_absolute_error(train_y,rf_train_pred))
print("RMSE: ",math.sqrt(mean_squared_error(train_y,rf_train_pred)))
```

# DecisionTree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import math
from math import *

DT = DecisionTreeRegressor()
DT.fit(train_X,train_y)
DT_pred = DT.predict(val_X)
DT_train_pred = DT.predict(train_X)
print("-"*10,"Test Pred","-"*10)
print("R2: ",r2_score(val_y,DT_pred))
print("MAE: ",mean_absolute_error(val_y,DT_pred))
print("RMSE: ",math.sqrt(mean_squared_error(val_y,DT_pred)))

print("-"*10,"Train Pred","-"*10)
print("R2: ",r2_score(train_y,DT_train_pred))
print("MAE: ",mean_absolute_error(train_y,DT_train_pred))
print("RMSE: ",math.sqrt(mean_squared_error(train_y,DT_train_pred)))
```
